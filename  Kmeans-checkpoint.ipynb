```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

dataset = pd.read_csv("C:/Users/gamle/Downloads/d.csv")

np.random.seed(97)

def seuil(data):
    matrix = data.select_dtypes(include=['int', 'float']).values
    dist_matrix = np.sqrt(np.sum((matrix[:, None] - matrix) ** 2, axis=2))
    mean_dist = np.mean(dist_matrix)
    std_dist = np.std(dist_matrix)

    seuil = mean_dist + 2 * std_dist
    return seuil*1.2

def distance(x, y):
    num_dims = []
    cat_dims = []
    for col in range(len(x)):
        if isinstance(x[col], (int, float)):
            num_dims.append(col)
        else:
            cat_dims.append(col)

    # Calculating euclidean distance for numeric dimensions
    num_dist = np.linalg.norm(x[num_dims] - y[num_dims])

    # Calculating categorical distance for categorical dimensions
    cat_dist = sum([x[cat_dim] != y[cat_dim] for cat_dim in cat_dims])/len(cat_dims)

    # Calculating total distance
    dist = np.sqrt(num_dist ** 2 + cat_dist ** 2)
    return dist

def optimize_distance(cluster, point, threshold):
    current_distance = np.mean([distance(p, point) for p in cluster])
    new_cluster = cluster.copy()
    new_cluster.append(point)
    new_distance = np.mean([distance(p, point) for p in new_cluster])
    if (new_distance - current_distance) / current_distance > threshold:
        return False
    else:
        return True

def create_group(point):
    return [point]

def algorithme_iteratif(data):
    data = pd.DataFrame(data)
    center = data.values[np.random.choice(data.shape[0])]
    grps = []
    grp = create_group(center)

    for i in data.values:
        d = distance(center, i)
        if d <= seuil(data):
            if optimize_distance(grp, i, seuil(data)):
                grp.append(i)
            else:
                grps.append(grp)
                grp = create_group(i)
        else:
            grps.append(grp)
            grp = create_group(i)
    grps.append(grp)
    return len(grps)



def kmeans(X, k):
    enc = OrdinalEncoder()
    scaler = StandardScaler()
    X = pd.DataFrame(X)
    X_num = X.select_dtypes(include=['int', 'float'])
    X_cat = X.select_dtypes(include=['object', 'category'])
    if not X_num.empty:
        X_num_scaled = scaler.fit_transform(X_num)
        X[X_num.columns] = X_num_scaled
    if not X_cat.empty:
        X_cat_encoded = pd.get_dummies(X_cat)
        X = pd.concat([X_num, X_cat_encoded], axis=1)
    X = X.values
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]
    while True:
        distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
        labels = np.argmin(distances, axis=0)
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    return centroids, labels

k = algorithme_iteratif(dataset.values)
print('The optimal number of clusters is:', k)

centroids, labels = kmeans(dataset, k)
dataset['label'] = labels

enc = OrdinalEncoder()
dataset = pd.DataFrame(enc.fit_transform(dataset), columns=dataset.columns)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(dataset)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.title('K-means clustering')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```
